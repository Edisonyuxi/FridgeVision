```python
# Step 4: Train EfficientNet Model with Optuna
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def evaluate_model_and_plot_random_correct_examples(model, loader, device, num_samples=3):
    model.eval()
    all_preds, all_labels = [], []
    correct_examples = []

    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            for i in range(len(images)):
                if preds[i] == labels[i]:
                    correct_examples.append((images[i].cpu(), labels[i].cpu(), preds[i].cpu()))

    if len(correct_examples) > num_samples:
        correct_examples = random.sample(correct_examples, num_samples)

    if correct_examples:
        fig, axes = plt.subplots(1, len(correct_examples), figsize=(15, 5))
        for i, (image, label, pred) in enumerate(correct_examples):
            image = image.permute(1, 2, 0)
            image = image * torch.tensor([0.229, 0.224, 0.225]) + torch.tensor([0.485, 0.456, 0.406])
            image = image.numpy().clip(0, 1)
            axes[i].imshow(image)
            axes[i].axis('off')
            axes[i].set_title(
                f'Label: {test_data.classes[label.item()]}\nPredicted: {test_data.classes[pred.item()]}',
                fontsize=12
            )
        plt.tight_layout()
        plt.show()

    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=1)
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=1)
    f1 = f1_score(all_labels, all_preds, average='weighted', zero_division=1)

    return accuracy, precision, recall, f1

def objective(trial):
    learning_rate = trial.suggest_float("learning_rate", 1e-5, 1e-2, log=True)
    batch_size = trial.suggest_categorical("batch_size", [32, 64, 128])
    epochs = trial.suggest_int("epochs", 5, 10)

    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False)

    model = load_efficientnet_model(num_classes).to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    accuracy, _, _, _ = evaluate_model_and_plot_random_correct_examples(model, test_loader, device)
    return accuracy

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=5)

best_params = study.best_params
print("Best hyperparameters:", best_params)

batch_size = best_params["batch_size"]
epochs = best_params["epochs"]
learning_rate = best_params["learning_rate"]

train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False)

model = load_efficientnet_model(num_classes).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()

for epoch in range(epochs):
    model.train()
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

torch.save(model, "/content/efficientnet_trained.pth")
print("EfficientNet model saved!")

accuracy, precision, recall, f1 = evaluate_model_and_plot_random_correct_examples(model, test_loader, device)
print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
```

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load YOLO model
yolo_model_path = "/content/yolov8n.pt"  # Replace with your YOLO model path
yolo_model = YOLO(yolo_model_path)

# Function to perform YOLO detection and plot results
def plot_yolo_detections(image_path, yolo_model):
    # Perform YOLO detection
    detections = yolo_model(image_path)

    # Load the image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR (OpenCV) to RGB (Matplotlib)

    # Plot the image with bounding boxes
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')

    for result in detections:
        boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding box coordinates
        labels = result.boxes.cls.cpu().numpy()  # Get class labels
        confidences = result.boxes.conf.cpu().numpy()  # Get confidence scores

        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box)
            label = result.names[int(labels[i])]
            confidence = confidences[i]

            # Draw the bounding box
            plt.gca().add_patch(
                plt.Rectangle((x1, y1), x2 - x1, y2 - y1, edgecolor='red', linewidth=2, fill=False)
            )
            # Add the label and confidence score
            plt.text(x1, y1 - 10, f"{label} ({confidence:.2f})", color='red', fontsize=12, backgroundcolor='white')

    plt.show()

# Example usage
image_path = "/content/image.jpg"  # Path to the input image
plot_yolo_detections(image_path, yolo_model)
```

```python
from ultralytics import YOLO
import cv2
from torchvision import transforms
from PIL import Image

# Load YOLO model
yolo_model_path = "/content/yolov8n.pt"  # Replace with your YOLO model path
yolo_model = YOLO(yolo_model_path)

# Load EfficientNet model
model_path = "/content/efficientnet_final.pth"
model = torch.load(model_path).to(device)
model.eval()

# Define transform for EfficientNet
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Function to integrate YOLO and EfficientNet
def detect_and_classify(image_path, cnn_model, yolo_model, class_names):
    # YOLO detection
    detections = yolo_model(image_path)
    crops = []
    yolo_detected_ingredients = []

    # Load image for cropping
    img = cv2.imread(image_path)
    for result in detections:
        boxes = result.boxes.xyxy.cpu().numpy()  # Bounding box coordinates
        labels = result.boxes.cls.cpu().numpy()  # YOLO class labels
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box)
            crop = img[y1:y2, x1:x2]
            crop_pil = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
            crops.append(crop_pil)
            yolo_detected_ingredients.append(result.names[int(labels[i])])

    # EfficientNet classification
    cnn_classified_ingredients = []
    for crop in crops:
        input_tensor = transform(crop).unsqueeze(0).to(device)
        with torch.no_grad():
            output = cnn_model(input_tensor)
            _, predicted_label = torch.max(output, 1)
            cnn_classified_ingredients.append(class_names[predicted_label.item()])

    # Combine YOLO and CNN results
    final_ingredients = list(set(yolo_detected_ingredients + cnn_classified_ingredients))
    return yolo_detected_ingredients, cnn_classified_ingredients, final_ingredients

# Example usage
image_path = "/content/test.jpg"
class_names = train_data.classes  # Class names from EfficientNet training
yolo_detected, cnn_classified, final_output = detect_and_classify(image_path, model, yolo_model, class_names)

# Print results
print("YOLO detected ingredients:", yolo_detected)
print("EfficientNet classified ingredients:", cnn_classified)
print("Final combined output:", final_output)
```

```python
import cv2
import torch
from torchvision import transforms
from PIL import Image
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load YOLO model for cropping
yolo_model_path = "/content/yolov8n.pt"  # Replace with your YOLO model path
yolo_model = YOLO(yolo_model_path)

# Load EfficientNet CNN model for classification
cnn_model_path = "/content/efficientnet_final.pth"  # Replace with your EfficientNet model path
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
cnn_model = torch.load(cnn_model_path).to(device)
cnn_model.eval()

# Define transform for EfficientNet input
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Function to crop food regions using YOLO and classify them using EfficientNet
def crop_and_classify_with_confidence(image_path, yolo_model, cnn_model, class_names):
    # Perform YOLO detection
    detections = yolo_model(image_path)

    # Load the image
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

    cropped_ingredients = []  # To store crops for visualization
    classified_labels = []    # To store classified labels
    confidences = []          # To store confidence scores

    for result in detections:
        boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding box coordinates

        # Loop through detected boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box)  # Extract box coordinates
            crop = img_rgb[y1:y2, x1:x2]   # Crop the region from the original image
            crop_pil = Image.fromarray(crop)  # Convert to PIL Image
            cropped_ingredients.append(crop_pil)  # Save the cropped image for visualization

            # Preprocess the cropped image for CNN
            input_tensor = transform(crop_pil).unsqueeze(0).to(device)

            # Classify using EfficientNet
            with torch.no_grad():
                output = cnn_model(input_tensor)
                probabilities = torch.nn.functional.softmax(output, dim=1)  # Get probabilities
                confidence, predicted_label = torch.max(probabilities, 1)  # Get confidence and label
                classified_labels.append(class_names[predicted_label.item()])
                confidences.append(confidence.item())

    return cropped_ingredients, classified_labels, confidences

# Function to visualize cropped ingredients, their classifications, and confidence scores
def visualize_crops_labels_and_confidence(cropped_ingredients, classified_labels, confidences):
    plt.figure(figsize=(15, 5))
    for i, (crop, label, confidence) in enumerate(zip(cropped_ingredients, classified_labels, confidences)):
        plt.subplot(1, len(cropped_ingredients), i + 1)
        plt.imshow(crop)
        plt.title(f"{label}\nConf: {confidence:.2f}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Example usage
image_path = "/content/test.jpg"  # Path to the input image
class_names = train_data.classes  # Class names from EfficientNet training
cropped_ingredients, classified_labels, confidences = crop_and_classify_with_confidence(image_path, yolo_model, cnn_model, class_names)

# Visualize the results with confidence scores
visualize_crops_labels_and_confidence(cropped_ingredients, classified_labels, confidences)

# Print classified labels with confidence scores
for label, confidence in zip(classified_labels, confidences):
    print(f"Classified: {label}, Confidence: {confidence:.2f}")
```
